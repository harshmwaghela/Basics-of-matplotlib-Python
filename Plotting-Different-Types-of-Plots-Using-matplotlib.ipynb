# 1.1 Simple plots

Let us start with the basic functions for plotting, .plot() and .show(). The below cell plots a point at (5,5). Also, the type of marker and the color is specified by the format string, ’*g’. plt.show() shows the output of the plot.

```python
plt.plot(5,5,"*g")
plt.show()
```

In the above example, only a single point has been plotted. If there are multiple points provided to plot(), then a line plot is generated as shown below. The format string ,‘o:k’ specifies the‘[marker][line-style][color]’ for the plot. The order may change in some places without affecting the output. ‘k’ stands for black color. Format strings are optional. For more in formation on format strings, click [here](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.plot.html).

```python
x= np.linspace(0,1,10)
y=2*x
plt.plot(x,y,'o:k') #marker, line, color / all color option: rgbrgbcmykw | all,linestyles '-', '--', '-.' ':'
```

# 1.2  Adding grids

Grids can be added to plots by setting plt.grid(True).

```python
x= np.linspace(0,1,10)
y=2*x
plt.grid(True)
plt.plot(x,y,'^:k') #marker, line, colour
plt.show()
```

# 1.3  Resizing plots

Plots can be resized to different figure sizes as per the requirement as shown below.

```python
x=np.linspace(0,1,10)
y=2*x
plt.figure(figsize=(10,7))
plt.plot(x,y,'^:k') #marker, line, colour
plt.show()
```

# 1.4  Line plots

Line plots can be plotted using .plot(). and providing the data for x and y axis as shown below.

```python
x=np.linspace(0,10,1000)
y=np.sin(x)
plt.plot(x,y,'o:r') #marker,line,color
plt.show()
```

# 1.5  Multiple lines on a single plot

One plot can have multiple lines as shown. By default, the lines will be plotted in different colors.

```python
x= np.linspace(0,10,1000)
y=np.sin(x)
z=np.cos(x)
plt.plot(x,y)
plt.plot(x,z)
plt.show()
```

# 1.6 Adding legends and using different markers and colours 

Plotting multiple plots on the same chart requires each plot to have a different color or line style. This can be done as shown. Also,adding legends helps to make such plots more read-able. Legends can be added by using plt.legend(). To read more about legends, click [here](https://jakevdp.github.io/PythonDataScienceHandbook/04.06-customizing-legends.html)

```python
x= np.linspace(0,10,100)
y=np.sin(x)
z=np.cos(x)
plt.figure(figsize=(20,10))
plt.plot(x,y,'*-r')
plt.plot(x,z,'o-b')
plt.legend(("SINUSIOD","COSINE"))
plt.grid(True)
plt.show()
```

Same plot has been shown below but ‘fmt’ has not been used here. Notice how specifications have been given using different parameters for color, linestyle and marker.

```python
x=np.linspace(0,10,100)
y=np.sin(x)
z=np.cos(x)
plt.figure(figsize=(20,10))
plt.plot(x,y,color='red',linestyle='solid', marker='*')
plt.plot(x,z,color='blue',linestyle='solid',marker='o')
plt.legend(("SINUSIOD","COSINE"))
plt.grid(True)
plt.show()
```

# 1.7  Changing the line width

Linewidth controls the thickness of lines and can be used as shown below.

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
plt.figure(figsize=(20,10))
plt.plot(x, y, '*-r', linewidth=5)
plt.plot(x, z, 'o-b')
plt.show()
```

# 1.8  Plot title and axis labels

Plot title and axis labels can be added to plots as shown below.

```python
x=np.linspace(0,10,30)
y=np.sin(x)
plt.figure(figsize=(10,7))
plt.plot(x, y,'^:g')#marker,line,color
plt.title('Sinusoidal Variations')
plt.xlabel('Variabale')
plt.ylabel('Sin of Variable')
plt.show()
```

# 1.9  Setting x and y axis limit

Limits can be set on x and y axis using .xlim() and .ylim() and specifying the range. The below graph is the same as the one shown above, except the range of the axis.

```python
x= np.linspace(0,10,30)
y=np.sin(x)
plt.figure(figsize=(10,7))
plt.xlim(-1,15)
plt.ylim(-2,2)
plt.plot(x,y,'^:g')
plt.show()
```

# 1.10  Creating subplots

Subplots can be created to show multiple plots in the same figure window.

```python
plt.subplot(2,1,1) #(rows,columns,panel number)
plt.plot(x, np.sin(x))
plt.subplot(2,1,2) #(rows,columns,panel number)
plt.plot(x, np.cos(x))
plt.show()
```

The below code will create a figure window having two rows and two columns.

```python
plt.subplot(2,2,1)# (rows, columns, panel number)
plt.plot(x, np.sin(x))

plt.subplot(2,2,2)
plt.plot(x, np.cos(x))

plt.subplot(2,2,3)
plt.plot(x, np.exp(x))

plt.subplot(2,2,4)
plt.plot(x, np.tan(x))

plt.show()
```

# 1.11  Adding annotations

Annotations may be added to plots to mark important points as shown below. The annotation text, point to annotate and location of annotation is inputted to annotate(). The color of the arrowcan be specified using arrow props as shown.

```python
t= np.arange(0.0,5.0,0.01)
s= np.cos(2*np.pi*t)
plt.plot(t,s, lw=2)
plt.annotate("Important",xy=(2,1),xytext=(3,1.5),
            arrowprops=dict(facecolor="black"))
plt.ylim(-2,2)
plt.grid(True)
plt.show()
```

# 1.12  Publishing plots

Plots created using matplotlib can be saved in various formats for further use in presentations,reports or documents or for sharing with other people. plt.savefig(‘Filename’) will save the file in the current directory. There are many formats to choose from such as png, jpg, pdf, svg and eps.The format can be given as an extension to the file name. For example, the cell shown below will automatically save the plot in png format with the name ‘my_figure.png’. If the filename is given as ‘my_figure.svg’, then the file will be saved in svg format.

eps and pdf are some of the widely used formats due to good quality images with small file size.

```python
x=np.linspace(0,10,100)
y=np.sin(x)
z=np.cos(x)
plt.plot(x,y,"r--",linewidth=5)
plt.plot(x,z,"b")
plt.savefig("my_figure.png")
```

# 1.12.1 Plot quality and background

The plot quality can be controlled by setting the ‘dpi’ parameter. For good quality plots, dpi canbe set to 300. Plots can be saved with transparent background by setting ‘transparent= True’ in.savefig().

```python
x=np.linspace(0,10,100)
y=np.sin(x)
z=np.cos(x)
plt.plot(x,y,"r--",linewidth=5)
plt.plot(x,z,"b")
plt.savefig("subplot.eps",format="eps", dpi=1200, transparent=True)
```

# 2  Various types of plots

Matplotlib can plot various types of plots. Let us start with the scatter plot.

# 2.0.1 Scatter plot

To plot a scatter plot, .scatter() is used as shown.

```python
x=np.linspace(0,5,5)
y=np.sqrt(x)
plt.scatter(x,y)
```

Let us try this with the iris dataset Iris dataset is loaded from sklearn library.iris.data will have 4 features, out of which only 2 features are taken for the present purpose. iris.target will have the group number for each of the three flower species.

```python
from sklearn import datasets
iris= datasets.load_iris()
data=iris.data[:,:2] #we only take the first two features
group= iris.target
```

A scatter plot is shown below where the first feature is plotted against the second feature. The color of each point is determined by the group to which it belongs, which is given by 'c=group'. 's=100' controls the size of the marker. cmap is used for color combinations. For more information, click [here](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html)

```python
plt.figure(figsize=(10,7))
plt.xlabel("Sepalwidth")
plt.ylabel("Sepallength")
plt.title("Irisdataset")
plt.scatter(data[:,0],data[:,1],s=100,c=group,cmap='viridis')
plt.show()
```

Other parameters such as 'alpha' to control transparancy of markers, 'cmap' to determine the color map and 'edgecolor' for markers can be set as shown.

```python
plt.figure(figsize=(10,7))
plt.scatter(data[:,0], data[:,1], s=200, c=group, 
            cmap='plasma', alpha=0.5,edgecolor='k')
plt.show()
```

# 2.0.2 Histogram

A histogram can be plotted using .hist() as shown below. It shows the frequency of 0 and 1 in data, which is 3 and 5 respectively.

```python
a= [1]*5 #a=[1,1,1,1,1]
b=[0]*3 #b=[0,0,0]
data=a+b #[1,1,1,1,1,0,0,0]
plt.hist(data)
```

The histogram below shows the frequency of random numbers within each interval.Number of bins and `edgecolor' have been set to specific values.

```python
data=np.random.randn(1000) #generates 1000 random values
plt.hist(data, bins="auto", edgecolor="white")
```

# 2.0.3 Horizontal Histogram

Histograms can also be plotted horizontally by setting the orientation=`horizontal'.

```python
data=np.random.randn(1000)
plt.hist(data,edgecolor="white",orientation="horizontal")
plt.show()
```

# 2.0.4 Bar plot

A bar plot can be plotted for categorical data as shown in the below examples.

Example 1 Simple bar plot

```python
shop= ["Electronics","Grocery","Apparel"]
customers=[50,125,43]
plt.bar(shop, customers)
plt.show()
```

Example 2 Using custom colors

```python
month= ["Jan","Feb","Mar","Apr","May","June",
        "July","Aug","Sep","Oct","Nov","Dec"]
sales=[500,600,800,700,300,250,100,150,600,700,1000,750]
plt.figure(figsize=(10,7))
plt.bar(month, sales, color=(0.0,0.5,0.4),edgecolor="black") #color can be specified as a combination of RGB.
plt.show()
```

Example 3 Multiple bar plots

```python
month=["Jan", "Feb","Mar","Apr","May","June",
       "July","Aug","Sep","Oct","Nov","Dec"]
sales_electronics= [500,600,800,700,300,250,100,150,600,700,900,750]
sales_apparel=[400,200,300,100,87,45,56,100,150,200,430,160]
plt.figure(figsize=(15,9))
plt.bar(month,sales_electronics,color=(0.69,0.25,0.96,0.7),
        edgecolor="black") #RGB and transperancy
plt.bar(month,sales_apparel,color=(0.30,0.25,0.96,0.9),
        edgecolor="black") #RGB and transperancy
plt.xlabel("Months")
plt.ylabel("Sales")
plt.legend(("Electronics","Apparel"),loc="upper right")
plt.show()
```

Example 4 Horizontal bar plot

```python
cake=["vanilla","chocolate","pineapple"]
popularity=[26,24,50]
plt.barh(cake,popularity,color=["blue","green","orange"],edgecolor="black")
plt.show()
```

# 2.0.5 Box plot

A box plot can be plotted as shown below for the iris data set. The orange lines show the median values. The horizontal box edges show first and third quartile.

```python
data=iris.data
plt.figure(figsize=(10,7))
plt.boxplot(data,labels=["Sapal Length","Sapal Width","Petal Length","Petal Width"])
plt.show
```

# 2.0.6 Violin plot

Violin plot have been shown for the same iris dataset. The data distribution has been shown for each of the 4 features. This is an advantage of a violin plot overthe box plot.

```python
plt.figure(figsize=(10,7))
plt.violinplot(data,showmedians=True)
plt.show()
```
